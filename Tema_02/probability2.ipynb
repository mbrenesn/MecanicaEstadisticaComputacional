### Desigualdad de Chebyshev y la ley de los grandes números

Empezamos evaluando la **desigualdad de Markov**.

**Proposición**: **Desigualdad de Markov**

Si $X$ es una variable aleatoria que toma solamente valores no negativos, entonces
$$
P\{ X \geq a \} \leq \frac{E[X]}{a}.
$$

---
Para la prueba, podemos definir una variable aleatoria $Y$ tal que
$$
Y = \begin{cases} a,\;{\textrm{si}}\; X \geq a \\ 0,\;{\textrm{si}}\; X < a \end{cases}.
$$
Dado que $X \geq 0$, se cumple que
$$
X \geq Y.
$$
Podemos evaluar el valor de expectación de la desigualdad anterior, luego
$$
E[X] \geq E[Y] = aP\{X \geq a \}
$$
<div style="text-align: right;">
$\blacksquare$
</div>

**Corolario**: **Desigualdad de Chebyshev**

Un corolario de la desigualdad de Markov corresponde a la desigualdad de Chebyshev. 

Si $X$ es una variable aleatoria que posee valor promedio $\mu$ y varianza $\sigma^2$, entonces para un valor $k > 0$ se cumple que
$$
P\{ |X - \mu| \geq k\sigma \} \leq \frac{1}{k^2}.
$$
Esto implica que la probabilidad de que una variable aleatoria difiera de su valor medio por una cantidad de $k$ veces su desviación estándar **está acotada** por un valor $1 / k^2$.

Recordamos que la desviación estándar corresponde a la raíz cuadrada de la varianza.

---
Para la prueba, sabemos que la variable aleatoria $(X - \mu) / \sigma^2$ debe ser no negativa y su valor de expectación debe ser
$$
E\Bigg[\frac{(X - \mu)^2}{\sigma^2}\Bigg] = \frac{E[(X - \mu)^2]}{\sigma^2} = 1.
$$
Ahora bien, de la desigualdad de Markov, se sigue que
$$
P\Bigg\{ \frac{(X - \mu)^2}{\sigma^2} \geq k^2 \Bigg\} \leq \frac{1}{k^2}.
$$
La desigualdad de Chebyshev sigue dado que la desigualdad $\frac{(X - \mu)^2}{\sigma^2} \geq k^2$ es equivalente a la desigualdad $|X - \mu| \geq k\sigma$.
<div style="text-align: right;">
$\blacksquare$
</div>

#### Teorema: La ley débil de los grandes números

Sea $X_1, X_2, \cdots$ una *secuencia* de variables aleatorias **independientes e idénticamente distribuidas** (**I.I.D**), cada una con valor medio $\mu$. Entonces, para cualquier valor $\epsilon > 0$,
$$
P\Bigg\{ \Bigg| \frac{X_1 + \cdots + X_n}{n} - \mu \Bigg| \geq \epsilon \Bigg\} \to 0,\;{\textrm{para}}\;n \to \infty.
$$
Esta ley es de suma importancia en la toería de probabilidad. Fundamentalmente, la probabilidad de que el **promedio** (ojo, el promedio no es el valor de expectación) de $n$ términos en una secuencia de variables aleatorias I.I.D difiera de su valor de expectación por más de un valor $\epsilon$, **tiende a cero cuando $n$ tiende a infinito**. 

---
Para la prueba debemos realizar la suposición adicional de que las variables aleatorias $X_i$ tienen una varianza finita $\sigma^2$. Luego
$$
E\Bigg[ \frac{X_1 + \cdots + X_n}{n} \Bigg] = \frac{E[X_1] + \cdots + E[X_n]}{n} = \mu,
$$
además
$$
{\textrm{var}}\Bigg[ \frac{X_1 + \cdots + X_n}{n} \Bigg] = \frac{{\textrm{var}}[X_1] + \cdots + {\textrm{var}}[X_n]}{n^2} = \frac{\sigma^2}{n}.
$$
Note que en la última expresión usamos el hecho de que las variables son I.I.D tal que la covarianza es cero. Con estos resultados, sustituimos en la desigualdad de Chebyshev de la siguiente forma:
$$
X := \frac{X_1 + \cdots + X_n}{n},
$$
de manera tal que usando los resultados anteriores tenemos:
$$
\mu \to \mu,\; \sigma \to \frac{\sigma}{\sqrt{n}}.
$$
Luego
$$
P\Bigg\{ \Bigg| \frac{X_1 + \cdots + X_n}{n} - \mu \Bigg| \geq \frac{k\sigma}{\sqrt{n}} \Bigg\} \leq \frac{1}{k^2}.
$$
Entonces, para cualquier $\epsilon > 0$ podemos escoger $\epsilon = k\sigma / \sqrt{n}$; o, de forma equivalente, 
$$
k^2 = \frac{n\epsilon^2}{\sigma^2}.
$$
Con lo cual obtenemos
$$
P\Bigg\{ \Bigg| \frac{X_1 + \cdots + X_n}{n} - \mu \Bigg| \geq \epsilon \Bigg\} \leq \frac{\sigma^2}{n\epsilon^2}.
$$
Note que en el límite de cuando $n \to \infty$, el lado derecho de la ecuación anterior tiende a cero.
<div style="text-align: right;">
$\blacksquare$
</div>

#### La ley fuerte de los grandes números

La ley débil de los grandes números se puede generalizar a la ley fuerte de los grandes números. Dicha ley establece que con probabilidad 1 se cumple que
$$
\lim_{n \to \infty} \frac{X_1 + \cdots + X_n}{n} = \mu.
$$
Es un resultado muy importante que el **promedio** (ojo de nuevo; el promedio, no el valor de expectación) de variables aleatorias I.I.D converge al valor de expectación en el límite de muchas variables.

### Variables aleatorias discretas

Existen diversas variables aleatorias que aparecen frecuentemente en ciertas aplicaciones. Haremos un resumen de algunas de ellas.

#### Variables aleatorias binomiales

Suponga que tenemos $n$ pruebas independientes, cada una con una probabilidad $p$ de "éxito". Si $X$ representa la cantidad de éxitos que ocurren para las $n$ pruebas, entonces decimos que $X$ **es una variable aleatoria binomial con parámetros $(n, p)$**. La PMF de la distribución tiene la forma
$$
P_i := P\{ X = i \} = \binom{n}{i} p^i (1 - p)^{n - i},\; i = 0, 1, \cdots, n
$$
donde
$$
\binom{n}{i} := \frac{n!}{i! (n - i)!}
$$
corresponde al coeficiente binomial, equivalente a la cantidad de subconjuntos de tamaño $i$ que pueden ser escogidos de un conjunto de $n$ elementos. El caso particular de una variable aletoria binomial con $n = 1$ se conoce como **variable aleatoria de Bernoulli.** 

**Valor de expectación y varianza**

Dado que la variable $X$ representa el número de éxitos en $n$ pruebas independientes, cada una con probabilidad $p$, podemos representarla de la forma
$$
X = \sum_{i = 1}^n X_i
$$
donde 
$$
X_i = \begin{cases} 1\; {\textrm{si i-ésima prueba es un éxito}} \\ 0\; {\textrm{de lo contrario}} \end{cases}
$$
Luego
$$
E[X_i] = P\{X_i = 1\} = p.
$$
Además,
\begin{align}
{\textrm{var}}[X_i] &= E[X_i^2] - (E[X_i])^2 \\
&= p - p^2 = p(1 - p).
\end{align}
Para este último resultado usamos $X_i^2 = X_i$, dado que $0^2 = 0$ y $1^2 = 1$. Luego, para la variable binomial $(n, p)$
$$
E[X] = \sum_{i=1}^n E[X_i] = np.
$$
Adicionalmente, dado que los $X_i$ son independientes
$$
{\textrm{var}}[X] = \sum_{i=1}^n {\textrm{var}}[X_i] = np(1-p). 
$$

#### Variables aleatorias de Poisson

Una variable aleatoria $X$ que toma los valores 0, 1, 2, ... se llama variable aleatoria de Poisson con parámetro $\lambda$, $\lambda > 0$, si su PMF está dada por
$$
p_i := P\{ X = i \} = e^{-\lambda} \frac{\lambda^i}{i!},\; i = 0, 1, \cdots
$$
Las variables aleatorias de Poisson son muy frecuentes. La razón se debe a que se utilizan para aproximar distribuciones del número de éxitos de un número muy grande de pruebas cuando la probabilidad de éxito es muy baja. 

De hecho, se puede probar que la variable aleatoria de Poisson es una variable aleatoria binomial en el límite de cuando $n$ es muy grande y $p$ muy pequeño (probar esto).

**Valor de expectación y varianza**

Conociendo el resultado anterior, podemos calcular los parámetros de la distribución siguiendo el resultado de la variable aleatoria binomial. Definimos $\lambda = np$, luego
$$
E[X] = np = \lambda
$$
y
$$
{\textrm{var}}[X] = np(1 - p) \approx np = \lambda.
$$
Entonces
$$
E[X] = {\textrm{var}}[X] = \lambda.
$$

#### Variable aleatoria geométrica

Considere pruebas independientes, cada una con una probabilidad $p$ de éxito. Si $X$ representa el número de pruebas hasta alcanzar una que sea exitosa, entonces
$$
P\{ X = n \} = p(1 - p)^{n-1},\; n \geq 1
$$
Lo cual se puede notar facilmente dado que las primeras $n-1$ pruebas son fallidas, hasta alcanzar la prueba $n$ la cual es exitosa. La PMF anterior se conoce como la PMF de la variable aleatoria geométrica con parámetro $p$.

**Valor de expectación y varianza**

Tenemos que
$$
E[X] = \sum_{n=1}^{\infty} np(1-p)^{n-1}.
$$
Esta suma la podemos evaluar usando
$$
\sum_{n=1}^{\infty} nx^{n-1} = \frac{d}{dx} \Bigg( \sum_{n=0}^{\infty} x^n \Bigg) = \frac{d}{dx} \Bigg( \frac{1}{1-x} \Bigg) = \frac{1}{(1-x)^2}
$$
Lo cual implica que
$$
E[X] = \frac{1}{p}.
$$
No realizaremos la prueba (una forma de hacerlo es sumando y restando $E[X]$ a la expresión de la varianza), pero se puede obtener que
$$
{\textrm{var}}[X] = \frac{1-p}{p^2}.
$$

### Variables aleatorias continuas

#### Variable aleatoria distribuida uniformemente

Una variable aleatoria $X$ se dice que está distribuida uniformemente si en el intervalo $(a,b)$, con $a < b$, la PDF está dada por
$$
f(x) = \begin{cases} \frac{1}{b-a}\;{\textrm{si}}\; a < x < b \\ 0\;{\textrm{de lo contrario}} \end{cases}
$$
Esta variable aleatoria será muy importante dado que podremos generar otras distribuciones con base a esta.

**Valor de expectación y varianza**

Tenemos
$$
E[X] = \frac{1}{b-a}\int_a^b x{\textrm{d}}x = \frac{b+a}{2}
$$
y
$$
E[X^2] = \frac{1}{b-a}\int_a^b x^2{\textrm{d}}x = \frac{a^2 + b^2 + ab}{3}.
$$
Entonces
$$
{\textrm{var}}[X] = \frac{1}{12} (b-a)^2.
$$
Además, la CDF corresponde a
$$
F(x) = \{X \leq x \} = \int_a^b (b-a)^{-1} {\textrm{d}}x = \frac{x-a}{b-a}. 
$$

#### Variable aleatoria normal

Una variable aleatoria se dice que está distribuida de forma normal si
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-(x-\mu)^2/2\sigma^2},\; -\infty < x < \infty.
$$
La variable normal tiene forma de campana centrada en $\mu$. 

**Valor de expectación y varianza**

No es díficil demostrar que
$$
E[X] = \mu
$$
y
$$
{\textrm{var}}[X] = \sigma^2.
$$
Es importante que si $X$ es una variable aleatoria normal con valor de expectación $\mu$ y varianza $\sigma^2$, entonces $aX + b$ está distribuida normalmente con valor de expectación $a\mu + b$ y varianza $a^2 \sigma^2$. Se sigue que si $X$ es normal con parámetros de la distribución $\mu, \sigma^2$, entonces
$$
Z = \frac{X - \mu}{\sigma}
$$
es normal con valor de expectación $0$ y varianza $1$.

### Teorema del límite central

La utilidad de las variables aleatorias normales surge de uno de los teoremas más importantes de la teoría de probabilidad. 

**Teorema**: **Límite central**

Sean $X_1, X_2, \cdots$ una secuencia de variables aleatorias I.I.D con valor de expectación finito $\mu$ y varianza finita $\sigma^2$. Entonces
$$
\lim_{n \to \infty} P\Bigg\{ \frac{X_1 + \cdots X_n - n\mu}{\sigma\sqrt{n}} < x \Bigg\} = \Phi(x),
$$
donde 
$$
\Phi(x) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{x} e^{y^2 / 2} {\textrm{d}}y,\; -\infty < x < \infty
$$

#### Variable aleatoria exponencial

La variable aleatoria exponencial es de suma importancia en la mecánica estadística. Esto se debe a que ciertas variables termodinámicas corresponden a variables aleatorias exponenciales. La PDF de la distribución corresponde a
$$
f(x) = \lambda e^{-\lambda x},\; 0 < x < \infty.
$$
La CDF está dada por 
$$
F(x) = \int_{0}^{x} \lambda e^{-\lambda y} {\textrm{d}}y = 1 - e^{-\lambda x},\; 0 < x < \infty
$$

**Valor de expectación y varianza**

Por medio de integración se puede probar que
$$
E[X] = \frac{1}{\lambda}
$$
y
$$
{\textrm{var}}[X] = \frac{1}{\lambda^2}.
$$

#### Variable aleatoria gamma

La variable aleatoria gamma tiene la PDF
$$
f(t) = \lambda e^{-\lambda t} \frac{(\lambda t)^{n-1}}{(n-1)!},\; t > 0
$$
con los parámetros $(n, \lambda)$. 

Un resultado muy importante corresponde a que **la suma de $n$ variables aleatorias exponenciales I.I.D. con parámetro $\lambda$, corresponde a una variable gamma con parámetros $(n, \lambda)$.**
