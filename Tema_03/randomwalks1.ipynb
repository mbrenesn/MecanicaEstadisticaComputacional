## Caminos aleatorios y propiedades emergentes

Las leyes matemáticas que describen el mundo macroscópico son, en general, simples.
- Se podría pensar que en el transfondo de lo que percibimos existen procesos sumamente complejos; sin embargo, las leyes fundamentales de los macroscópico siguen siendo reglas relativamente sencillas.

- **¿A qué se debe esto? ¿Cuales son los mecanismos sobre los cuales las leyes macroscópicas poseen la capacidad de predicción, a pesar de que los procesos microscópicos son tan complejos?**

En la mecánica estadística, hablamos de una teoría que identifica la razón por la cual los mecanismos macroscópicos **emergen** de la **complejidad subyacente**.

### El movimiento Browniano

El movimiento Browniano corresponde al **movimiento irregular de partículas microscópicas** (tales como perfume en un medio aéreo o el polen) suspendidas en un fluido.
- Este comportamiento fue observado por Robert Brown en 1827 usando un microscopio.
    - El comportamiento, a como fue descrito, persiste inclusive en **partículas inertes**, i.e., el efecto es **físico**, no biológico.

#### Teoría de Einstein (1905)
<small>
A. Einstein, Über die von der molekularkinetischen Theorie der Wärme geforderte Bewegung von in ruhenden Flüssigkeiten suspendierten Teilchen.

Annalen der Physik, 322(8), 549–560 (1905)
</small>

Einstein, en uno de sus tres célebres trabajos de 1905, explicó su trabajo seminal sobre el movimiento Browniano **usando física estadística**:
- Una partícula microscópica es bombardeada por incontables moléculas.
- Las colisiones asociadas son **aleatorias y sin correlaciones**, dando lugar a un *desplazamiento neto aleatorio*.

Estudiaremos la (importante) derivación en la siguiente lección, sin embargo, un importante resultado corresponde a
$$
E[ X^2(t) ] = 2 D t,
$$
donde $D$ es una constante conocida como **coeficiente de difusión**. En la expresión anterior:
- $X$ es una variable aleatoria que representa la posición de una partícula microscópica suspendida en un fluido
    - La posición cambia con el tiempo, lo cual implica que **tenemos que estudiar variables aleatorias como función de un parámetro temporal**.
    - El valor de expectación de la posición al cuadrado escala linealmente con el tiempo. 

Este resultado es revolucionario por tres razones:
- Corresponde a uno de los primeros resultados importante en física estadística.
- Conecta fluctuaciones microscópicas (varianza, distribuciones probabilísticas) con movimiento macroscópico.
- Y, tal vez lo más importante, predice de forma téorica lo observado experimentalmente con respecto a la **existencia de átomos**.

#### Experimentos de Perrin (1908–1909)
<small>
Perrin, J., Mouvement brownien et réalité moléculaire.

Annales de Chimie et de Physique, 18, 5–114 (1909)
</small>

Jean Perrin estudió la difusión de partículas coloidales:
- Obtuvo excelente concordancia de los resultados con la predicción de Einstein.
- Corresponde a uno de las primeras confirmaciones cuantitativas de la teoría atómica.

Estos resultados son tan importantes (dependiendo de el valor que usted le de a esto) que le dieron a Perrin el premio Nobel en Física en 1926.

#### Caminos aleatorios

Los caminos aleatorios son el *pilar* de una teoría estadística para describir fenómenos físicos, con los cuales podemos estudiar:
- Corrientes  y cantidades conservadas
- Difusión
- *Respuesta lineal* (esto es importante)
- Distribuciones de Boltzmann

Cabe destacar que todos estos resultados **parten de primeros principios** usando nada más que teoría probabilística, sin aludir a las grandes teorías de la Física.

La idea del camino aleatorio corresponde **a considerar la posición de una partícula (objeto, átomo, molécula, almendras, papayas, el valor de cierta acción en la bolsa de valores) como una variable aleatoria que depende de un parámetro, en este caso, el tiempo**.  

#### Procesos estocásticos

¿Recuerda nuestra definición de proceso estocástico? (Tema_01/random.ipynb)

Básicamente, un proceso estocástico es uno en el cual se modela una *cantidad física como una variable aleatoria que varía con algún parámetro*.
- Los caminos aleatorios son nuestro primer proceso estocástico
- Piense en las implicaciones. Ya conocemos la teoría fundamental de probabilidad; ahora los valores de expectación, varianzas, PDFs y otros conceptos fundamentales de la teoría **pasan a ser funciones del tiempo**.

En muchos casos, podemos considerar **versiones discretas de este parámetro**, mientras que en otras debemos considerar su espectro contínuo.

#### Camino aleatorio simétrico en 1D

Un camino aleatorio corresponde al proceso estocástico en el cual un objeto que parte del origen, se puede mover a la izquierda o a la derecha con **la misma probabilidad** en cada paso. 

Podemos parametrizar el tiempo en el espacio discreto usando una colección de $N$ variables aleatorias $\xi_i$, donde $\xi$ es $\pm 1$ para el $i$-ésimo paso. De esta forma
$$
X(N) = \sum_{i=1}^{N} \xi_i,
$$
donde las $N$ variables aleatorias $\xi$ son I.I.D con PMF
$$
P\{ \xi = l \} = \frac{1}{2},\; l = +1, -1.
$$

Note que hemos discretizado el "tiempo", de manera tal que en cada unidad de tiempo se da un "paso". Esto define un proceso de Markov en espacio-tiempo discreto, pero todavía no entraremos en esos detalles.

##### Valor de expectación

¿Cúal será el valor de expectación $E[X(N)]$? Veamos.

Por la propiedad fundamental del valor de expectación, tenemos
$$
E[X(N)] = \sum_{i=1}^N E[\xi_i].
$$
A su vez,
$$
E[\xi_i] = (+1)\cdot\frac{1}{2} + (-1)\cdot\frac{1}{2} = 0,
$$
lo cual implica 
$$
\boxed{E[X(N)] = 0.}
$$
Físicamente, el desplazamiento **medio** en un camino aleatorio es cero. El objeto, en promedio, no se desplaza.

##### Varianza.

La varianza es más interesante. Veamos
\begin{align}
{\textrm{var}}[X(N)] &= E[X^2(N)] - \big(E[X(N)]\big)^2 \\
&= E[X^2(N)].
\end{align}
Luego
\begin{align}
X^2(N) &= \left(\sum_{i=1}^N \xi_i \right)^2 = \sum_{i=1}^N \xi_i^2 + 2\sum_{1 \le i < j \le N} \xi_i \xi_j \\
\implies E[X^2(N)] &= E\left[ \left(\sum_{i=1}^N \xi_i \right)^2 \right] = E \left[ \sum_{i=1}^N \xi_i^2 \right] + 2 E\left[ \sum_{1 \le i < j \le N} \xi_i \xi_j \right]
\end{align}
Recordemos que para dos variables aleatorias $X$ y $Y$ I.I.D., $E[XY] = E[X] E[Y]$ (Tema_02/probability1.ipynb), lo cual implica que el último término en la ecuación anterior
$$
E\left[ \sum_{1 \le i < j \le N} \xi_i \xi_j \right] = 0.
$$
Luego
$$
E \left[ \sum_{i=1}^N \xi_i^2 \right] = \sum_{i=1}^N E[\xi_i^2] = \sum_{i=1}^N 1 = N,
$$
por lo tanto
$$
\boxed{{\textrm{var}}[X(N)] = N.}
$$

Este resultado es **importante**. En general, en física no se suele hablar de la varianza, si no de su raíz cuadrada. El término que se suele utilizar es la *raíz cuadrática media* del desplazamiento:
$$
\sigma = \sqrt{{\textrm{var}}[X(N)]} = \sqrt{N},
$$
la cual indica que la variación del desplazamiento con respecto al promedio es $\sqrt{N}$ ($L\sqrt{N}$ para pasos de tamaño $L$, en lugar de 1). 
- Esto es un resultado **sumamente interesante**, implica que a pesar de que el desplazamiento en promedio es cero, sus fluctuaciones **son finitas**.
- Existen varias formas de derivar este resultado. Una forma es considerando el camino aleatorio como una distribución binomial, o de forma inductiva a-la Capítulo 2.1 Sethna.
- Estas fluctuaciones **tienen una consecuencia importante a nivel físico**, e.g., el movimiento Browniano y la ecuación de difusión.

#### Desde la perspectiva de la distribución binomial

Visto desde otra perspectiva, los caminos aleatorios nos darán un resultado sumamente importante de algunos procesos estocásticos: la universalidad.

Para empezar, consideramos un camino aleatorio unidimensional de $N$ pasos. Para cada paso:
- La probabilidad de un paso a la derecha +1: $p = 1/2$
- La probabilidad de un paso a la izquierda -1: $q = 1 - p = 1/2$

De manera tal que para una secuencia de $N$ pasos, donde $k$ son los pasos a la derecha, tenemos la probabilidad de la **secuencia**
$$
p^{k} q^{N - k} = \left( \frac{1}{2} \right)^{k} \left( \frac{1}{2} \right)^{N - k} = \left( \frac{1}{2} \right)^{N},
$$
de manera tal que para **cualquier secuencia** de $N$ pasos tiene **probabilidad equivalente $2^{-N}$**.

¿Qué pasa si contamos cuántas secuencias dan lugar a $k$ pasos a la derecha?

La cantidad de secuencias que tienen $k$ pasos a la derecha y $N - k$ pasos a la izquierda **corresponde al número de posibilidades de escoger $k$ de un total $N$ de secuencias** (*$N$ choose $k$*). Esto corresponde al coeficiente binomial
$$
\binom{N}{k}.
$$

Entonces, ¿cuál es la probabilidad total de tener un camino aleatorio con $k$ pasos a la derecha? Basta con multiplicar la cantidad de secuencias que dan lugar a este desplazamiento, multiplicada por la probabilidad de cada una de las secuencias. Esto da lugar a la **PMF de la distribución de los caminos aleatorios unidimensionales**:
$$
P\{ K = k \} = \binom{N}{k} \left( \frac{1}{2} \right)^N,\;k = 0, 1, \cdots, N
$$
**la cual corresponde a la distribución binomial $K \sim {\textrm{binom}}(N, p = 1/2)$**.

**Importante:** Note que esta no es la distribución de $X(N)$ como teníamos anteriormente, corresponde a la distribución de dar $k$ pasos a la derecha, la cual es un poco diferente.

#### Intuición

Considere $N = 3$. Para este caso, tenemos solamente $2^3 = 8$ posibles combinaciones: DDD, DDI, DID, IDD, III, IID, IDI e IID (I = izquierda, D = derecha). La probabilidad de cada secuencia es $2^{-3} = 1 / 8$.
- Entonces; para, e.g., $k = 2$ derechas, existen las posibilidades DDI, IDD y DID, i.e., $\binom{3}{2} = 3$ secuencias cada una con probabilidad $1 / 8$. Luego, la probabilidad de desplazarse dos espacios a la derecha es $3 / 8$.
<div style="text-align: right;">
$\blacksquare$
</div>

De esta forma, la distribución binomial surge naturalmente para los caminos aleatorios.
Para calcular el valor de expectación y la varianza, podemos calcular el desplzamiento con respecto a esta distribución. Para eso, tenemos
$$
X(N) = (+1) \cdot k + (-1) \cdot (N - k) = 2k - N.
$$
Luego,
\begin{align}
E[X(N)] &= \sum_{k=0}^N (2k - N) P\{ K = k \} \\ 
&= 2\sum_{k=0}^N k P\{ K = k \} - \sum_{k=0}^N N P\{ K = k \} \\
&= 2\sum_{k=0}^N k P\{ K = k \} - N \sum_{k=0}^N P\{ K = k \} \\
&= 2\sum_{k=0}^N k P\{ K = k \} - N.
\end{align}
Además, sabemos que el primer término al lado derecho de la última ecuación corresponde al valor de expectación de una variable $K$ distribuida bajo la distribución binomial, i.e., $E[K] = N p$, donde $p = 1/2$. Entonces
$$
\boxed{E[X(N)] = 2 \cdot N \cdot \frac{1}{2} - N = 0}.
$$
Este es el mismo resultado que obtuvimos anteriormente.

Para la varianza, tenemos que
$$
X(N) = 2k - N,
$$
donde $k$ es una variable distribuida de forma binomial. Recordamos que
$$
{\textrm{var}}[aK + b] = a^2{\textrm{var}}[K], 
$$
donde, en este caso, tenemos $a = 2$ y $b = -N$. Dado que var$[K] = Np(1 - p)$ para la binomial, con $p = 1/2$, entonces
$$
\boxed{{\textrm{var}}[X(N)] = a^2 {\textrm{var}}[K] = 2^2 \cdot N \cdot \frac{1}{2} \left( 1 - \frac{1}{2} \right) = N}.
$$
Igual que nuestro resultado anterior.

### Universalidad e invarianza de escala

Los caminos aleatorios poseen una propiedad muy importante de la física estadística: la invarianza de escala.

En general, hemos demostrado que un camino de $N$ pasos se desplaza, en promedio, $\sqrt{N}$. Los caminos aleatorios:
- Forman trayectorias entrelazadas, de manera tal que si hacemos "zoom in" a un camino en particular para $N$ grande, la versión aumentada tiene la misma forma que el anterior.
- A este tipo de estructuras, en matemáticas, se les conoce como **fractales**.
- Siempre y cuando $N$ sea lo suficientemente grande como para no poder distinguir los detalles pequeños del camino, cada camino aleatorio es **autosimilar**.

Una forma sencilla de visualizar caminos aleatorios usando `Numpy` corresponde a usar la función `np.random.choice`. Para dos dimensiones podemos hacer:

```python
import numpy as np
import matplotlib.pyplot as plt

NPasos = 2000
NCaminos = 5

pasosX = np.random.choice([-1, 1], size = (NCaminos, NPasos))
pasosY = np.random.choice([-1, 1], size = (NCaminos, NPasos))
```

Las posiciones totales en cada dirección se pueden evaluar sumando cada uno de los pasos aleatorios, para esto podemos usar `np.cumsum`. Veamos:

```python
xPositions = pasosX.cumsum(axis = 1)
yPositions = pasosY.cumsum(axis = 1)
```

Ya tenemos arreglos que contienen las trayectorias. Solo falta visualizar:

```python
plt.figure(figsize=(6,6))
for i in range(NCaminos):
    plt.plot(xPositions[i], yPositions[i], alpha=0.6)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Movimiento Browniano simulado en 2D")
plt.axis("equal")
plt.show()
```

La universalidad se refiere al siguiente concepto:
- **Sobre escalas en las cuales los pasos individuales no son discernibles y las correlaciones entre pasos son de igual forma muy pequeñas para ser detectadas, todos los caminos aleatorios se ven de la misma forma.**

Este fenómeno se refiere a que, estadísticamente, los detalles microscópicos *no importan*. Colisiones de fotones, tiros aleatorios de monedas, moléculas de perfume en un fluido; todos estos son ejemplos sobre los cuales el mismo modelo aplica.

Esta propiedad es más que una observación fenomenológica y se puede **demostrar matemáticamente**.

Los detalles microscópicos dejan de ser relevantes, en este modelo, cuando tenemos el caso de $N$ grande. Lo importante de los caminos aleatorios es que poseen la propiedad de **varianza finita**:
- Esto implica que se puede aplicar el **teorema del límite central**.

Veamos como aplicar el teorema del límite central en este caso. Recordemos que $X(N) = 2k - N$. Entonces
$$
P\{ K = k \} = \binom{N}{k} \, 2^{-N}
$$
y
$$
P\{ X(N) = x \} = \binom{N}{\tfrac{N+x}{2}} \, 2^{-N}.
$$
La fórmula de Stirling para un valor grande de $m$ es tal que
$$
m! \;\approx\; \sqrt{2 \pi m}\; m^m e^{-m}.
$$
Entonces
$$
\binom{N}{k} \;\approx\;\frac{\sqrt{2\pi N}\, N^N e^{-N}} {\sqrt{2\pi k}\, k^k e^{-k}\, \sqrt{2\pi (N-k)} \,(N-k)^{N-k} e^{-(N-k)}}.
$$
Los factores $e^{-N}$, $e^{-k}$ y $e^{-(N-k)}$ se cancelan, luego:
$$
\binom{N}{k} \;\approx\; \frac{1}{\sqrt{2\pi}} \sqrt{\frac{N}{k(N-k)}} \cdot \frac{N^N}{k^k (N-k)^{N-k}}.
$$
Para la distribución de $X(N)$, tenemos $k = \tfrac{N+x}{2}$. Entonces, definimos
$$
p = \frac{k}{N} = \frac{1}{2} + \frac{x}{2N}, \qquad q = 1-p = \frac{1}{2} - \frac{x}{2N}.
$$
Entonces
$$
\binom{N}{k} \;\approx\; \frac{1}{\sqrt{2\pi N p q}} \;\frac{1}{p^{Np} q^{Nq}}.
$$
Por ende, la PMF corresponde a
$$
P\{K = k \} \approx \frac{1}{\sqrt{2\pi N p q}} \left( \frac{1}{2^N p^{Np} q^{Nq}} \right).
$$
Ahora, note que $p \approx q \approx \frac{1}{2}$ para $x / N$ pequeño, lo cual implica que podemos hacer una expansión.

Sea $\epsilon = \tfrac{x}{2N}$, entonces $p = \tfrac{1}{2} + \epsilon$. La expansión de $\ln p$ alrededor de $1 / 2$ es  
$$
\ln\!\left(\tfrac{1}{2}+\epsilon\right) = \ln\!\tfrac{1}{2} + 2\epsilon - 2\epsilon^2 + O(\epsilon^3).
$$
Similarmente
$$
\ln\!\left(\tfrac{1}{2}-\epsilon\right) = \ln\!\tfrac{1}{2} - 2\epsilon - 2\epsilon^2 + O(\epsilon^3).
$$
Podemos valernos de estas aproximaciones si calculamos el logaritmo de PMF y luego la elevamos con la función exponencial:
$$
P\{ K = k \} \;\approx\; \frac{1}{\sqrt{2\pi N p q}} \exp\!\Big[-N\big(p \ln \tfrac{p}{1/2} + q \ln \tfrac{q}{1/2}\big)\Big].
$$
Entonces, para el exponente
$$
-N \Big[ p \ln\!\frac{p}{1/2} + q \ln\!\frac{q}{1/2} \Big] = -N \Big[ p(\ln p - \ln \tfrac{1}{2}) + q(\ln q - \ln \tfrac{1}{2}) \Big].
$$
Usando la expansión anterior, tenemos
\begin{align}
-N \Big[ p(\ln p - \ln \tfrac{1}{2}) + q(\ln q - \ln \tfrac{1}{2}) \Big] &= -N \Big[ p(2\epsilon - 2\epsilon^2) + q(-2\epsilon - 2\epsilon^2) \Big] \\
&= -N \Big[ (\frac{1}{2} + \epsilon)(2\epsilon - 2\epsilon^2) + (\frac{1}{2} - \epsilon)(-2\epsilon - 2\epsilon^2) \Big] \\
&= -2N\epsilon^2.
\end{align}
Además,
$$
-N(2\epsilon^2) = -N \cdot 2 \left(\frac{x^2}{4N^2}\right) = -\frac{x^2}{2N}.
$$
Finalmente, (haciendo $p \approx q \approx 1/2$ y sabiendo $X(N) = 2k - N$)
$$
\boxed{P\{ X(N) = x \} \;\approx\; \frac{1}{\sqrt{2\pi N}} \exp\left(-\frac{x^2}{2N}\right)}
$$

Esto corresponde a una distribución Gaussiana $\mathcal{N}(0, N)$. De esta manera, la **condición de universalidad es una consecuencia directa del teorema del límite central**.

La teoría de caminos aleatorios descrita anteriormente se generaliza directamente a más de una dimensión, siempre y cuando los pasos en cada dirección correspondan a variables aleatorias I.I.D.
